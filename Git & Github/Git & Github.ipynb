## What is Git?
- Git is a version control system used to manage and control the versions of the softwares, programs we use in our project.
- Using git we can connect from the local repository(pc) to the remote repository(github) and we can upload and save all the changes we make in our project in local repository to the remote repository.

## What is version control?
- As the name suggests version control is controlling and managing the versions of the softwares, programs we use in our project.
- Version control is a way to manage different versions of the same software we use.

## Why version control?
1. The primary reason for version control is collaboration.
    - In real life industry, we work as a team in building a project. All the members of the team has to integrate, connect, share their work without affecting the work of others. Therefore version control is a way to collaborate teams in a project.
2. Maintaining the version.
    - While building a project we face with situations where the different versions of the same softwares in different stages of the project will lead to errors. therefore we have to use same versions of the software in different stages of the project and version control is a way to do the same.

## Git
- Distributed version control system.
- Local Repository/system = Git
- Remote Repository/system = Github
  
  

## 1. Git snapshot:
- A snapshot or picture in git is a represenation of the current state of the project or code at a given time.
- Snapshots are synonymous with commits. Every time you commit to a Git repository, you are saving a snapshot of all the files in your repository.

## 2. Git status:
- Every time we make a change or modify a file or project there are 3 stages in which git saves it.
- 1. <font color = blue>Modified</font> - The file has been modified since last commit.
  2. <font color = blue>Staged</font> - The file is ready to be saved.
  3. <font color = blue>Commited</font> - The file is saved.
 
## 3.Implementation:
- In command prompt change the directory as per requirements.
- 1. To initialize git - <font color = blue>**git init**</font>
  2. To check status - <font color = blue>**git status**</font>
  3. To add git - <font color = blue>**git add.**</font>
  4. To commit - <font color = blue>**git commit -m "........"**</font>
  5. To check past commits - <font color = blue>**git log**</font> [which ever is the last commit it will be called as HEAD]
