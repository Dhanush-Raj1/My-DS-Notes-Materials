# 8. Measure of Asymmetry
- Skewness is the most common way of measuring asymmetry.
- skewness indicates that whether the data is concentrated on one side or not.

- It is difficult to find skewness with formula therefore use python or excel.


- By looking at the image it is clear that
   1. **In a left skewed distribution:**
      - Mean < median < mode.
      - Majority of the data points lies at the peak with lower positive values and a fewer data points lies at the tail with negative values.
      - Examples:
         - Price of luxury goods. 
           - Only a fewer number of luxury goods are available at a lower price whereas higher number of luxury goods are at a higher price.   
         - Test scores of a easy exam.
           - Only a fewer students would score a low marks in a easy test whereas majority of the students will score higher marks.
         - Product lifespan.
           - Only a few number of products tend to have a short lifespan whereas majority of products will have a longer life span. 
&nbsp;    
   2. **In a symmetrical distribution:**
      - Mean=Median=Mode.
      - Data points lie equally on the right and the left of the distribution.  
&nbsp;    
   3. **In a right skewed distribution:**
      - Mean > median > mode
      - Majority of the data points lies at the peak with lower values and a fewer data points lies at the tail with higher values.
      - Examples: 
        - Wealth Distribution.
          - Majority of the people will have less amount of wealth whereas only a few number of people will tend to have more amount of wealth.
        - Test scores of a difficult exam.
          - More number of students tend to score lesser marks whereas only a few students would score higher marks.
        - Household Income.
          - Majortiy of the people's household income is lower whereas minority of the people's household income is higher.

````python
## function for deteting outliers using zscore and threshold of 3 std
def detect_outliers(data):
    outliers = []
    threshold = 3
    mean = np.mean(data)
    std = np.std(data)
    
    for i in data:
        zscore = i-mean/std
        if np.abs(zscore)>threshold:
            outliers.append(i)
            
    return outliers
````

````python
datas = [1,2,3,4,5,20]
detect_outliers(datas)        ## asper zscore 4,5,20 are outliers
````

````python
## using IQR
def detect_outliers1(data):
    data_sort = sorted(data)
    outliers = []
    Q3 = np.percentile(data_sort,75)
    Q1 = np.percentile(data_sort,25)
    IQR = Q3-Q1
    Low_fence = Q1-(1.5*IQR)
    High_fence = Q3+(1.5*IQR)
    print(Low_fence, High_fence)
    
    for i in data_sort:
        if i < Low_fence or i > High_fence:
            outliers.append(i)
                    
    return outliers
````

# 11. Types of Distributions

## 11.1 Gaussian Distribution
- The Gaussian distribution also known as normal distribution is a continuous probability distribution for a random variable where the datapoints clusters around the mean.
- Gaussian distribution has a bell curve structure.
- In a Gaussian distribution the dataset is equally distribution i.e 50% on the right and 50 % on the left side.
- In a normal distribution the mean, median, mode will be on the centre of the distribution.
- Emperical rule or the 68-95-99.7% rule applies for the normal distribution.


## Empricial Rule:
- It is a guideline that describes the behavior of data in a normal distribution. It provides a rough estimate of the proportion of data that falls within a certain number of standard deviations from the mean.
- According to the empirical rule:
     - Approximately 68% of the data falls within one standard deviation of the mean.
     - Approximately 95% of the data falls within two standard deviations of the mean.
     - Approximately 99.7% of the data falls within three standard deviations of the mean.

## 11.1.1 Standard Normal Distribution
- Standard normal distribution is a simplest form of normal distribution with a μ=0 and σ=1.
- Also known as z-distribution.
## Z Score:
- Z score is used to determine at which standard deviation a specific data point falls from the mean.
- By applying the z score formula to the dataset we get the zscore values i.e z distribution.
- 
- The z score values we get from applying the zscore on a dataset, will be in **standard normal distribution** with a μ=0 and σ=1. These values (zscores) forms the z distribution.
  
### Applications:
1. In the data science project life cycle during the data pre processing/data engineering stage we perform feature scaling processes which includes standardization and normalization. Standardization  converts the dataset's μ into 0 and σ into 1 inorder to standardize the data.

- A QQ plot or Quantile Quantile is a graphical tool used to check whether a sample dataset follows a theoretical distribution usually standard normal distribution.
- QQ plot is plotted by 
     - Determining the quantiles of each data points in the sample dataset.
     - Determining the quantiles of each data point in a theoretical distribution (SND).
     - Plotting the quantiles of the sample dataset on the y axis.
     - Plotting the quantiles of the theoretical distribution in the x axis.
- **If the datapoints of the sample dataset approximately falls on the identity line of the theoretical distribution then the sample dataset is normally distributed.**

### Uses:
- QQ plots are used to compare the distribution of a sample dataset with a theoretical distribution.
- It is also used to determine which theoretical distribution does the sample dataset follows. 

````python
# necessary library
import statsmodels.api as sm

# extracting the feature from the dataset    
item_mrp = df['Item_MRP']

# plotting the qqplot
## This function takes an array or series of data as input and plots the quantiles of the data against the quantiles of a 
## specified theoretical distribution. By default, it uses the standard normal distribution as the theoretical distribution.
sm.qqplot(item_mrp, line='s')

plt.show()                           
````

**Conclusion: It is clear that the datapoints does not fall on the identity line of the theoretical distribution. Therefore there is not similarity between the two datasets and the "Item_MRP" feature does not follow a SND.**

# 14. Central Limit Theorem
- Central limit theorem is a fundamental concept which states that for a random variable X of any distribution, if i take multiple random sample datasets from the variable (100 sample datasets with each having a 100 data points) and if i plot the means of those 100 samples it will follow a normal distribution with the mean approximately equal to the mean of the variable X and the variance will be std/n.

### Conditions:
- The random variables in the sample should be independent and identically distributed.
- The size of each samples should be sufficiently large, typically considered to be 30 or more.

### Uses:
- It allows us to make inferences about the population mean based on the sample mean, even when the population distribution is unknown or non-normal.
- Base of hypothesis testing. 
- Base of confidence interval.

# 15. Chebyshev's Inequality
- Chebyshev's Inequality is a probability theory which is similar to the Empricial Rule except it is for a non Gaussian Distribution.
- Chebyshev's Inequality states that for a random variable X which belongs to a non Gaussian Distribution, what will be the probabilty for data points that falls at
    - μ + 1σ
    - μ + 2σ
    - μ + 3σ
- The probability for a random variable X where X is not a Gaussian distribution with some mean, std  $[X ≠ GD(μ,σ)]$    
     > $ Pr[μ - Kσ < x < μ + Kσ] > 1 - 1/K^2 $    
Where K is the value of standard deviation.      

- Suppose I need to find the probability of data points that falls between μ and 1σ, 
- **μ + 1σ**    
> $ Pr[μ - 1σ < x < μ + 1σ] > 1 - 1/1^2 = 0 $    
     
- **μ + 2σ**
> $ Pr[μ - 2σ < x < μ + 2σ] > 1 - 1/2^2 = 3/4 = 0.75 $ &nbsp;&nbsp;&nbsp;&nbsp;<u>[75% of the data points falls on μ + 2σ]<u>  
     
- **μ + 3σ**
> $ Pr[μ - 3σ < x < μ + 3σ] > 1 - 1/3^2 = 8/9 = 0.89 $ &nbsp;&nbsp;&nbsp;&nbsp;<u>[89% of the data points falls on μ + 2σ]<u>

# 16. P value
- P value also knowna as probability value is a measure that helps in assessing the strenght of evidence against null hypothesis HO. It gives the probability value for HO to be true or false.
- In hypothesis testing, the p-value is compared to a predetermined significance level (often denoted as α) to make a decision about rejecting or failing to reject the null hypothesis. Usual α = 0.05
- If pvalue < 0.05 α, we reject the null hypothesis HO because there is only less than 5% of evidence for HO to be true.
- If pvalue > 0.05 α. we accept the null hypothesis HO because there is more than 5% of evidence for HO to be true.

# 17. Hypothesis Testing 
- Hypothesis testing is a statistical method used to draw inferences and conclusions about a population based on a sample data. It is done by formulating null hypothesis HO and alternative hypothesis HA and determining the evidence(pvalue) to determine which hypothesis is more supported.
- The pvalue is compared to a alpha or significance level which is usually 0.05.
    - If pvalue < 0.05 - we reject HO
    - If pvalue > 0.05 - we accept HO 
      
#### Note:
In python performing hypothesis testing with ztest, ttest or anyother tests is different from performing manually.   
- Read the documentation of the tests that you are performing before using it in python.
- we require a sample dataset in order to perform the test but while performing manually we only require mean of sample and population and std of population and sample size.

## Procedure  
  - 1.Form hypothesis HO and Ha.  
  - 2.State alpha value α.
  - 3.Define decision rule. (confidence interval values by refering to z,t,chi tables)
  - 4.Calculate test statistics.
  - 5.Form conclusion.

## 17.1 Z-test 
- Z-test is a statistical test used for hypothesis testing.
- It is used to find whether the mean of a sample is significantly different from the mean of the population.
- **It is used when the σ population standard deviation is given and when the sample size is equal to or greater than 30 (n>=30).**
- The Z statistics is a measure of how much standard deviation a datapoint is away from the mean.
  
### Interpretation:
- The Z statistics is compared with a critical value which is determined by using z table. (1-0.05/2) where 1 is area of the graph and 0.05 is the alpha value.
- If the zvalue is more than the critical value than we reject the HO.

#### Note:
1. Ztest function by default uses sample standard deviation (s) to calculate z statistic but as per the formula population standard deviation (σ) must be used. The sample standard deviation is used as a point estimate for the population standard deviation because ztest function assumes that the population std is unknown.

Problem 1.
  Suppose the IQ in a certain population is normally distributed with a mean of μ=100 and standard deviation of σ=15. A researcher wants to know if a new drug affects the IQ level of individuals so he recruits 20 individuals to experiment on them. - HO : The drug does not affect the intelligence of the individuals. or μ=100.
- H1 : The drug has an effect on the intelligence of the individuals. or μ!=100.
- Alhpa α=0.05

````python
from statsmodels.stats.weightstats import ztest

# sample data
sample1 = [88, 92, 94, 94, 96, 97, 97, 97, 99, 99, 105, 109, 109, 109, 110, 112, 112, 113, 114, 115]

# HO
null_value = 100

## The ztest function by default uses the sample standard deviation(s)to calculate z statistics or                   
z_statistic, p_value = ztest(sample1, value=null_value)            ## we can calculate manually
                                                             
# printing results
print("Z-statistic:", z_statistic)
print("P-value:", p_value)
````

- The 1st value is the ztest value (z statistic) and 2nd value is the p value.
- Our α=0.05 and p value=0.110, here the pvalue is greater than α (tvalue>alpha) therefore we accept the null hypothesis.

**Conclusion : Since the pvalue is greater than the alpha value of 0.05, we accept the HO.**

## 17.3 Chi-Square test
- Chi-square test is a statistical tool used to determine whether there is a significant association between categorical variables.
- It is used to find the relationship or dependency between two categorical variables.
- **It is used when there are two categorical variables.**
- Chi-square statistics is the measure of the difference between the observed and the expected frequencies. 

### Interpretation:
- The chi statistics has to compared with the critical value which is determined by using chi-sqaure table. The critical value has to be found with degrees of freedom and alpha value.

Problem 3. The average weights of children around the age of 8-14 is recorded during 2013. And after a gap of 10 years the averge weights of children around the age of 8-14 is recorded during 2023. Did the population distribution of age has changed in the last 10 years?

- HO : 2023 data meets the distribution of 2013 data. (OR) There is no change in the distribution.
- H1 : 2023 data does not meet the distribution of 2013 data. (OR) There is a change in the distribution.
- Alpha : 0.05

````python
from scipy.stats import chisquare 

# Observed frequencies
observed = np.array([15, 16, 15, 14, 16, 17, 19, 20, 22, 21, 22, 23, 24, 25, 26, 27, 28, 28, 29, 30])

# Expected frequencies (null hypothesis)
expected = np.array([17, 18, 17, 18, 17, 18, 19, 21, 22, 23, 24, 25, 24, 25, 22, 23, 18, 28, 28, 30])

# Performing chi-square test
chi2, p_value = chisquare(observed, expected, ddof=1)

# printing results
print("Chi-square statistic:", chi2)
print("P-value:", p_value)
````

**Conclusion: Since the P value is greater than the significance level of 0.05, we are accepting HO.**

## 17.4 ANOVA
- ANOVA is used to compare the means of two or more groups to determine if there are any statistically significant differences among them. The comparison is done by looking at the variation in the data.
- The variation in the data is divided into two parts:
    - Variation between the groups - difference between the means of the groups 
    - Variation within the groups - difference between the speard of each datapoints within the groups (variability).
- ANOVA is used when we need to compare two or more groups.
- It is also used when we need to compare a numerical feature and with a categorical feature where the categorical feature has more than two different categories. 
- **Fvalue is the ratio of between-group variation and within-group variation.**

### Interpretation:
- The fvalue has to be compared with the critical value which is determined by using the f table. The critical value has to found by using the degrees of freedom and alpha value.

````python
# comparing the means of three groups

import scipy.stats as stats

# 3 different groups
group1 = [1,2,3,4,5,6]
group2 = [2,2,3,7,9,10]
group3 = [3,1,5,8,12,16]

# determining the fvalue and pvalue
f_value, p_value = stats.f_oneway(group1, group2, group3)

# printing the results
print(f'F value:',f_value, f'P value:', p_value)
````

**Conclusion: The Pvalue is greater than the significant level of 0.05 therefore we are accepting HO.**

# 19. Type I and Type II Errors
- Type I and Type II errors are the two types of error that occures during hypothesis testing.
1. Outcome 1:  
 - We reject the null hypothesis in reality if the null hypothesis is False. <font color=Green>**Correct Decision**</font>.  
2. Outcome 2:       
 - We reject the null hypothesis in reality if the null hypothesis is True. <font color=red>**Wrong Decision**</font>.<font color=Lime>**Type I Error**</font>    
3. Outcome 3:
   - We accept the null hypothesis in reality if the null hypothesis is True. <font color=green>**Correct Decision**</font>.
4. Outcome 4:
   - We accept the null hypothesis in reality if the null hypothesis is False. <font color=red>**Wrong Decision**</font>.<font color=Lime>**Type II Error**</font>
