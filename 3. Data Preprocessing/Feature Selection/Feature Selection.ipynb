```python
# creating a data frame 

import pandas as pd

df = pd.DataFrame({'A':[1,2,4,1,2,4],
                   'B':[4,5,6,7,8,9],
                   'C':[0,0,0,0,0,0],
                   'D':[1,1,1,1,1,1]})
```

# 1. Variance Threshold
Feature selector that removes all low-variance features.

This feature selection algorithm looks only at the features (X), not the desired outputs (y), and can thus be used for unsupervised learning.

## Dropping constant features
- Removing those features that has only constant values or zero variance as it of no use for solving the problem statement.

```python
from sklearn.feature_selection import VarianceThreshold
varthes = VarianceThreshold(threshold = 0)
varthes.fit(df)
```

```python
varthes.get_support()
```

```python
constant_columns = [column for column in df.columns
                    if column not in df.columns[varthes.get_support()]]
print(constant_columns)
```

- A and B features are non zero variance features and therefore they are usefully in solving a problem statement.
- C and B features are zero variance features and therefore they can be removed.

```python
# removing zero variance features

df.drop(constant_columns, axis=1, inplace=True)
```

## Working with kaggle dataset

We are gonna play with 'Santander Customer Satisfaction' dataset from kaggle.  
There are two file 'train.csv' and 'test.csv' .  
The dataset contains totally 371 features so lets get to work and start selecting only the useful features.   
The dependent feature is 'TARGET'. Rest all are independent features.    

```python
# importing the dataset

df_train = pd.read_csv(r'F:\Data Science\Datasets\Santander Customer Satisfaction\train.csv')
df_train.head()
```

```python
df_train.shape
```

The dataset contains around 76,000 rows and 371 columns.

```python
# separating dependent and independent features

x = df_train.drop('TARGET', axis = 1)
y = df_train['TARGET']
```

```python
x.head()
```

```python
y.head()
```

```python
# train, test, split

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 40)
```

```python
# applying varians threshold

varthes.fit(x_train)
```

```python
varthes.get_support()
```

```python
# non constant features

sum(varthes.get_support())
```

There are 327 features that are non-constant from the total of 370 features.  
So 43 features are zero variance features.  

```python
# names of the zero variance / constant features

constant_columns1 = [columns for columns in x_train.columns
                     if columns not in x_train.columns[varthes.get_support()]]
constant_columns1
```

```python
print(len(constant_columns1))
```

We can see that there are totally 43 constant features and we have to remove those.

```python
# removing constant features

x_train.drop(constant_columns1, axis = 1)
```

Now we have totally 327 useful features for solving our problem statement.  
We can also change the variance threshold from 0 to 5 or 10, by doing so will reduce the total number of constant features.
