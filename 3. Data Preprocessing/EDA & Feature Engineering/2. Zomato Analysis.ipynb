# Attribute information
- Restaurant Id: Unique id of every restaurant across various cities of the world
- Restaurant Name: Name of the restaurant
- Country Code: Country in which restaurant is located
- City: City in which restaurant is located
- Address: Address of the restaurant
- Locality: Location in the city
- Locality Verbose: Detailed description of the locality
- Longitude: Longitude coordinate of the restaurant's location
- Latitude: Latitude coordinate of the restaurant's location
- Cuisines: Cuisines offered by the restaurant
- Average Cost for two: Cost for two people in different currencies ðŸ‘«
- Currency: Currency of the country
- Has Table booking: yes/no
- Has Online delivery: yes/ no
- Is delivering: yes/ no
- Switch to order menu: yes/no
- Price range: range of price of food
- Aggregate Rating: Average rating out of 5
- Rating color: depending upon the average rating color
- Rating text: text on the basis of rating of rating
- Votes: Number of ratings casted by people

[Source](https://www.kaggle.com/datasets/shrutimehta/zomato-restaurants-data)

- Zomato dataset already contains a columns named "Country Code" we just have to merge "Country-Code.xlsx" with the same "Country Code" column.

```python
final_df = pd.merge(df, df_country, on='Country Code', how='left')
final_df.head()
```

## Pie chart

```python
country_names = final_df.Country.value_counts().index
country_names
```

```python
country_values = final_df.Country.value_counts().values
country_values
```

```python
# plotting only top 3 countries that uses zomato

explode = (0, 0.1, 0.1)                                         # explode=(0,0.1,0.1) only "explode" the 2nd & 3rd slice
plt.pie(country_values[:3], labels=country_names[:3], autopct='%1.2f%%', explode=explode, 
        colors = ['blue', 'red', 'green'])    
```

```python
ratings = final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(
                                                                           columns={0:'Rating Count'})
ratings
```

- When rating is above 4.5 it indicates Excellent.
- When rating is above 4.0 and below 4.5 indicats Very Good.
- And from 3.5 to 3.9 indicates Good.
- Ratings from 2.5 to 2.9 indicates Average.    
- Ratings from 1.9 to 2.4 is Poor.  
&nbsp;      
- Majority of the users around 2148 users did not give their rating.  

```python
plt.figure(figsize=(12, 6))       # plt.rcParams['figure.figsize'] = (8,6) can also be used for modifying size
sns.barplot(x='Aggregate rating', y='Rating Count',hue='Rating color', data=ratings, 
                  palette=['blue', 'Red', 'Orange', 'Yellow', 'Green', 'Green'])
```

- hue parameter allows you to add additional categorical variable. The color of the bars will be based on this categorical variable. 
- pallete parameter is used to assign colors to the bars in relation to hue. Since the colors from hue does not match the rating color we use pallete.

**Obsevation :**  
    1. 0.0 rating or not rating has maximum rating count.  
    2. Majority of the ratings are from 2.6 to 4.0.  

```python
plt.rcParams['figure.figsize'] = (8,6)
sns.barplot(x='Rating color', y='Rating Count', data=ratings, palette=['blue', 'red', 'orange', 'yellow', 'green', 'green'])
```

- White has maximum rating count.
- Followed by tht yellow and orange has majority of rating count.
- Then Green and dark green and red.

```python
# count plot

sns.countplot(x='Rating color', data=ratings, palette=['blue', 'red', 'orange', 'yellow', 'green', 'green'])
```

- Don't get confused with the above countplot.
- In the above graph we are only counting the number of colors that are recorded in rating color feature. We are viualizing only the frequency of colors in rating color.
- Orange has the maximum rating count and then red followed by that yellow, green and white.
