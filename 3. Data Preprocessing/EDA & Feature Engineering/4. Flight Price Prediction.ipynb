```python
# importing train dataset

df_train = pd.read_excel(r'F:\Data Science\Datasets\Flight Price\Data_Train.xlsx')
df_train.head()
```

```python
# importing test dataset 

df_test = pd.read_excel(r'F:\Data Science\Datasets\Flight Price\Test_set.xlsx')
df_test.head()
```

# Feature engineering

```python
df_final.info()
```

We can see that 'Date_of_Journey' is in object type. we have to change to numerical type.     
We are doing this by create new columns for date, month, year separately.    

## Handling Date_of_Journey

```python
# creating separate columns for data, month, year
df_final['Date'] = df_final['Date_of_Journey'].str.split('/').str[0]
df_final['Month'] = df_final['Date_of_Journey'].str.split('/').str[1]
df_final['Year'] = df_final['Date_of_Journey'].str.split('/').str[2]
```

```python
# we can also use lambda function to do the above 
# df_final['Date'] = df_final['Date_of_Journey'].apply(lambda x:x.split('/')[0])
# df_final['Month'] = df_final['Date_of_Journey'].apply(lambda x:x.split('/')[1])
```

```python
df_final.head()
```

We can see that Date, Month, year columns are created.

```python
# now we can change the dtype of those columns

df_final['Date'] = df_final['Date'].astype(int)
df_final['Month'] = df_final['Month'].astype(int)
df_final['Year'] = df_final['Year'].astype(int)
```

```python
df_final.info()
```

The dtype of Date, Month, Year columns are now in int32.
Since we have separete columns for Date, Month, Year, we can now remove the 'Date_of_Journey' column as it of no use.

```python
# removing 'Date_of_Journey' column

df_final.drop('Date_of_Journey', axis=1, inplace=True)
df_final.head()
```

## Handling Arrival_Time

Similarly we can do the same for 'Arrival_Time'. We can create separate columns for arrival_hour and arrival_min by extracting only the required information. 
We can start by extracting the exact  arrival time i.e 01:10 and we can ignore 22 Mar.  

```python
# extracting only the arrival time and ignoring the day of arrival
# this time lets use lambda function

df_final['Arrival_Time'] = df_final['Arrival_Time'].apply(lambda x : x.split(' ')[0])
```

```python
df_final.head()
```

We can see that the infomation in 'Arrival_Time' column contain only the time of arrival and not the day, month.
Now we can early create separate columns for 'Arrival_Hour' and 'Arrival_Min'.

```python
# creating separate columns for arrival hour and minute

df_final['Arrival_Hour'] = df_final['Arrival_Time'].apply(lambda x : x.split(':')[0])
df_final['Arrival_Min'] = df_final['Arrival_Time'].apply(lambda x : x.split(':')[1])
```

```python
# now we change the dtype of 'Arrival_Hour' and 'Arrival_Min' columns

df_final['Arrival_Hour'] = df_final['Arrival_Hour'].astype(int)
df_final['Arrival_Min'] = df_final['Arrival_Min'].astype(int)    
```

```python
# deleting the original 'Arrival_Time' column

df_final.drop('Arrival_Time', axis=1, inplace=True)
df_final.head()
```

```python
df_final.info()
```

Now there are separate columns for Arrival_Hour and Arrival_Min with dtype as integer.

## Handling Dep_Time

We can do the same for 'Dep_time' column as well. Extracting the info, creating separate columns, changing the dtype and finally deleting the original 'Dep_time' column.

```python
# extracting departure hour, minute and storing it in separate columns

df_final['Dep_Hour'] = df_final['Dep_Time'].apply(lambda x : x.split(':')[0])
df_final['Dep_Min'] = df_final['Dep_Time'].apply(lambda x : x.split(':')[1])
```

'Dep_Hour' and 'Dep_Min' columns have been created with its respective values.  
Now we can change the dtype of those columns.  

```python
# changing the dtype of 'Dep_Hour' and 'Dep_Min' columns into integer

df_final['Dep_Hour'] = df_final['Dep_Hour'].astype(int)
df_final['Dep_Min'] = df_final['Dep_Min'].astype(int)
```

```python
# removing the 'Dep_Time' column

df_final.drop('Dep_Time', axis=1, inplace=True)
df_final.head()
```

The 'Dep_Time' column have been removed and there are separete columns for 'Dep_Hour' and 'Dep_Min' with dtype int.

## Handling Route and Total_Stops

Route column contains the details of the route from the departure point to the destination point. There is also another column 'Total_Stops' which contains information about the total number of stops from the departure point till the destination point.

So we can use any one of those features as both contains similar information. 
Since 'Route' has the names of the inbetween stops and we cannot properly convert the names of those stops into numerical, we are removing Route feature and we are gonna make use of 'Total_Stops' feature.

```python
df_final['Total_Stops'].unique()
```

There is one nan value, lets see that record.

```python
df_final[df_final['Total_Stops'].isnull()]
```

The above is the record which contains the null value.  
The source is Delhi and destination is Cochin. Assuming there are 1 stops between Delhi and Cochin we can fill the nan value with 1 stop.  

```python
# encoding the values of 'Total_Stops' feature

df_final['Total_Stops'] = df_final['Total_Stops'].map({'non-stop':0, '1 stop':1, '2 stops':2, '3 stops':3, '4 stops':4,
                                                        'nan':1})
```

```python
df_final.head()
```

We can see that the categorical values of 'Total_Stops' columns have converted to numerical values.
Since we are using 'Total_Stops' feature, we can remove the 'Route' feature.

```python
# removing 'Route' feature'

df_final.drop('Route', axis=1, inplace=True)
df_final.head()
```

## Handling Duration 

### Duration_Hour

```python
# creating a separate column for Duration hour

df_final['Duration_Hour'] = df_final['Duration'].str.split(' ').str[0].str.split('h').str[0]
```

```python
df_final.info()
```

```python
# converting 'Duration_Hour' into int dtype

df_final['Duration_Hour'] = df_final['Duration_Hour'].astype(int)
```

While converting 'Duration_Hour' into dtype int we faced an error which says that there is a value in 'Duration_Hour' viz '5m' as a result we cannot convert the feature into dtype int.

```python
# extracting the records which contains 'Duration_Hour' == '5m'

df_final[df_final['Duration_Hour']=='5m']
```

The above are those records that contains 'Duration_Hour' as '5m'.  
Funny as per the records it says that to travel from Mumbai to Hyderbad it only takes 5 minutes.  
We can only assume that it is an anomaly or incorrect information. Therefore we are removing those two records.  

```python
# removing the records which contains 'Duration_hour' as '5m' using the index number

df_final.drop(6474, axis=0, inplace=True)
df_final.drop(2660, axis=0, inplace=True)
```

```python
# converting dtype of 'Duration_Hour' into int

df_final['Duration_Hour'] = df_final['Duration_Hour'].astype(int)
```

### Duration_Min

```python
df_final['Duration_Min'] = df_final['Duration'].str.split(' ').str[1].head(5).str.split('m').str[0]
```

```python
df_final.info()
```

```python
df_final['Duration_Min'].value_counts()
```

We can see that 'Duration_Min' has only 8 records in total.  
We can either add the respectively 8 records 'Duration_Min' with 'Duration_Hour' and convert the entire column into minutes.  
Or we can just drop those 8 records and delete 'Duration_Min' column.

```python
# deleting 'Duration_Min' feature

df_final.drop('Duration_Min', axis=1, inplace=True)
```

We can also delete the 'Duration' feature as we created a new feature 'Duration_Hour'

```python
# deleting 'Duration' feature

df_final.drop('Duration', axis=1, inplace=True)
```

```python
df_final.info()
```

Now we are only left with categorial features viz 'Airline', 'Source', 'Destination', 'Additional_Info'.  
We have to encode these categorial features.  

## Handling Airline

```python
df_final['Airline'].value_counts()
```

### LabelEncoder

```python
# importing label encoder

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
df_final['Airline'] = le.fit_transform(df_final['Airline'])
df_final['Source'] = le.fit_transform(df_final['Source'])
df_final['Destination'] = le.fit_transform(df_final['Destination'])
df_final['Additional_Info'] = le.fit_transform(df_final['Additional_Info'])
```

```python
df_final.head()
```

```python
df_final.info()
```

We have used labelencoding technique for converting those categorical values.  
We can also use onehot encoding.  

### OneHot encoder

```python
from sklearn.preprocessing import OneHotEncoder
oe = OneHotEncoder()
```

```python
df_final['Airline'] = oe.fit_transform(np.array(df_final['Airline']).reshape(-1,1)) 
df_final['Source'] = oe.fit_transform(np.array(df_final['Source']).reshape(-1,1)) 
df_final['Destination'] = oe.fit_transform(np.array(df_final['Destination']).reshape(-1,1)) 
df_final['Additional_Info'] = oe.fit_transform(np.array(df_final['Additional_Info']).reshape(-1,1)) 
```

Since we are getting multiple errors we can use pd.get_dummies().  
The reason for the errors could be we have already applied labelencoder.  

```python
pd.get_dummies(df_final,columns=["Airline", "Source", "Destination", "Additional_Info"] ,drop_first = True).head()
```

We can see that pd.get_dummies have sucessfully executed and the above is the final dataframe.
